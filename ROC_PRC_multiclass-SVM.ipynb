# bin them by KBinsDiscretizer

```python
# Binarize the output
from sklearn.preprocessing import label_binarize

y = label_binarize(df['ground truth'], classes=[0, 1, 2])
n_classes = y.shape[1]
```

```python
X = df[['MCF10A score']]
```

```python
X.shape
```

```python
from sklearn.preprocessing import KBinsDiscretizer

enc = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='kmeans')

MCF10A_binned = enc.fit_transform(df[['MCF10A score']])
```

```python
MCF10A_score = label_binarize(MCF10A_binned, classes=[0, 1, 2])
```

```python
# calculate macro-average roc
def mac_avg_roc(y_test, y_score):
    from scipy import interp
    
    # first ompute ROC curves and ROC areas for each class
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
    
    # aggregate all false positive rates
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))

    # Then interpolate all ROC curves at these points    
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(n_classes):
        mean_tpr += interp(all_fpr, fpr[i], tpr[i])

    # Finally average it and compute AUC        
    mean_tpr /= n_classes
    
    fpr_macro = np.insert(all_fpr, 0, 0)
    tpr_macro = np.insert(mean_tpr, 0, 0)
    
    return fpr_macro, tpr_macro
```

```python
fpr_MCF10A, tpr_MCF10A = mac_avg_roc(y, MCF10A_score)
roc_auc_MCF10A = auc(fpr_MCF10A, tpr_MCF10A)
```

```python
fig, ax = plt.subplots()

ax.plot(fpr_MCF10A, tpr_MCF10A,
         label='AUC MCF10A (area = {0:0.2f})'.format(roc_auc_MCF10A),
         color='deeppink')

ax.plot([0, 1], [0, 1], color='navy', linestyle='--')

ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.set_title('macro-avg ROC curves')
legend = ax.legend(loc="lower right")

plt.show()
```

# bin them by manual thresholding

```python
def thresholding(upper, lower):
    score =[]
    for i in df1['MCF10A score']:
        if i>upper:
            score.append(2)
        elif i<lower:
            score.append(0)
        else:
            score.append(1)
    score_bi = label_binarize(score, classes=[0,1,2])
    return score_bi
```

```python
# Compute micro-average ROC curve and area
def miu_avg_roc(test, score):
    fpr, tpr, _ = roc_curve(test.ravel(), score.ravel())
    # .ravel() returns a contiguous flattened array
    roc_auc = auc(fpr, tpr)
    return fpr, tpr, roc_auc
```

```python
import matplotlib as mpl

def plot_roc(fpr, tpr, roc_auc, name):
    plt.plot(fpr, tpr, label='AUC, {0}={1:0.3f}'.format(name, roc_auc))
    plt.plot([0,1], [0,1], color='navy', linestyle='--')
    
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
#    plt.xlim([0.0, 1.0])
#    plt.ylim([0.0, 1.05])
#    plt.title('micro-avg ROC curves')
    plt.legend(loc="lower right")
```

```python
score46 = thresholding(0.6, 0.4)
score37 = thresholding(0.7, 0.3)
score28 = thresholding(0.8, 0.2)
score27 = thresholding(0.7, 0.2)
score47 = thresholding(0.7, 0.4)
score57 = thresholding(0.7, 0.5)
```

```python
fpr_46, tpr_46, roc_auc_46 = miu_avg_roc(y, score46)
fpr_37, tpr_37, roc_auc_37 = miu_avg_roc(y, score37)
fpr_28, tpr_28, roc_auc_28 = miu_avg_roc(y, score28)
fpr_27, tpr_27, roc_auc_27 = miu_avg_roc(y, score27)
fpr_47, tpr_47, roc_auc_47 = miu_avg_roc(y, score47)
fpr_57, tpr_57, roc_auc_57 = miu_avg_roc(y, score57)
```

```python
plot_roc(fpr_46, tpr_46, roc_auc_46, 'threshold 0.4/0.6')
plot_roc(fpr_37, tpr_37, roc_auc_37, 'threshold 0.3/0.7')
plot_roc(fpr_28, tpr_28, roc_auc_28, 'threshold 0.2/0.8')
plot_roc(fpr_27, tpr_27, roc_auc_27, 'threshold 0.2/0.7')
plot_roc(fpr_47, tpr_47, roc_auc_47, 'threshold 0.4/0.7')
plot_roc(fpr_57, tpr_57, roc_auc_57, 'threshold 0.5/0.7')

mpl.rcParams['pdf.fonttype'] = 42
plt.savefig('Output PDFs/ROC multiclass binning.pdf', transparent=True, dpi=200)

plt.show()
```
